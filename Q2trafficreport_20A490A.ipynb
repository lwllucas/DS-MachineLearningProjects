# 2. Extract and format the traffic incident data require for display in Singapore map

```
for items in ti_data.items():
  report = items[1]

ti_list = []
msg_list = []
lat_list = []
long_list = []

for reports in report:
  ti_list.append(reports)

for i in range(len(ti_list)):
  msg_list.append(ti_list[i]['Message'])
  lat_list.append(ti_list[i]['Latitude'])
  long_list.append(ti_list[i]['Longitude'])
  

print(lat_list)
print(long_list)
print(msg_list)
```

# 3. Python program request the traffic band URL and return the JSON data

```

url = "http://datamall2.mytransport.sg/ltaodataservice/TrafficSpeedBandsv2"

payload = {}
headers = {
  'AccountKey': 'Mi+hn6YERECkhXsJv8/BKg==',
  'accept': 'application/json'
}

response = requests.request("GET", url, headers=headers, data = payload)

tsb_data = response.json()
```

# 4. Extract and format the traffic band data require for display in Singapore map

```
for items in tsb_data.items():
  tsb_report = items[1]

tsb_list = []

info_list = []
loc_list = [] 
lat_tsb = []
long_tsb = []
speed_band = []
road_category = []
max_speed = []
min_speed = []
link_id = []

for reports in tsb_report:
  tsb_list.append(reports)

for i in range(len(tsb_list)):
  info_list.append(tsb_list[i]['RoadName'])
  loc_list.append(tsb_list[i]['Location'])
  lat_tsb.append(float(loc_list[i].split(" ")[0]))
  long_tsb.append(float(loc_list[i].split(" ")[1]))
  speed_band.append(int(tsb_list[i]['SpeedBand']))
  road_category.append(tsb_list[i]['RoadCategory'])
  max_speed.append(int(tsb_list[i]['MaximumSpeed']))
  min_speed.append(int(tsb_list[i]['MinimumSpeed']))
  link_id.append(tsb_list[i]["LinkID"])


print(lat_tsb)
print(long_tsb)
print(info_list)
print(loc_list)
print(speed_band)
print(road_category)
print(max_speed)
print(min_speed)
print(link_id)
```

# 5. Add the data into the display map with different markers to represent the traffic incident and traffic bands

#6. When clicking on the marker on the display map it should show the information of either traffic incident or traffic bands

- 1 – indicates speed range from 0 < 9    --- Red
- 2 – indicates speed range from 10 < 19  --- Green
- 3 – indicates speed range from 20 < 29  --- Blue
- 4 – indicates speed range from 30 < 39  --- Yellow
- 5 – indicates speed range from 40 < 49  --- Orange
- 6 – indicates speed range from 50 < 59  --- Pink
- 7 – indicates speed range from 60 < 69  --- Purple
- 8 – speed range from 70 or more.        --- Teal

```
import folium

color = ["red", "green", "blue", "yellow", "orange", "pink", "purple", "teal"]
# Coordinates of Singapore
m = folium.Map(location=[1.360270, 103.851959],
                          zoom_start=11.55,
                       width=1200, height=600)
fill = ""

for i in range(len(msg_list)):
  folium.Marker(
      location=[lat_list[i],long_list[i]], # coordinates for the marker (Earth Lab at CU Boulder)
      popup= msg_list[i], # pop-up label for the marker
      icon=folium.Icon(icon='fa-circle', icon_color="white", prefix='fa')
  ).add_to(m)

for i in range(len(info_list)):
  
  if speed_band[i] == 1:
    fill = color[0]
  elif speed_band[i] == 2:
    fill = color[1]
  elif speed_band[i] == 3:
    fill = color[2]
  elif speed_band[i] == 4:
    fill = color[3]
  elif speed_band[i] == 5:
    fill = color[4]
  elif speed_band[i] == 6:
    fill = color[5]
  elif speed_band[i] == 7:
    fill = color[6]
  else:
    fill = color[7]
      
  folium.CircleMarker(
      location=[lat_tsb[i],long_tsb[i]], # coordinates for the marker (Earth Lab at CU Boulder)
      popup= "LinkID: " + str(link_id[i]) + " \n" + info_list[i] + "\nMax Speed: " + str(max_speed[i]) + "\nMin Speed: " + str(min_speed[i]) + "\nRoad Category: " + road_category[i], # pop-up label for the marker
      fill_color= fill,
      fill_opacity= 0.8
  ).add_to(m)

m
```
